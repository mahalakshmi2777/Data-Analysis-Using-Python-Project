```python
import matplotlib.pyplot as plt
months = ["Jan","Feb","Mar","Apr","May"]
sales = [10000,12000,15000,13000,18000]
plt.plot(months,sales)
plt.title("Monthly sales")
plt.xlabel("Months")
plt.ylabel("Sales Amount")
plt.show()
```

```python
import matplotlib.pyplot as plt
students = ["A","B","C","D"]
marks = [89,98,75,95]
plt.bar(students,marks)
plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
import matplotlib.pyplot as plt
products = ["Laptop","Mobile","Tablet"]
sales = [50,58,39]
plt.bar(products,
        sales,
        color=["red","blue","green"])
plt.title("Product Sales")
plt.show()
```

```python
import matplotlib.pyplot as plt
months = ["July","August","september","October","November","December"]
sales = [4000,5000,6000,7000,8000,9000]
plt.figure(figsize=(12,5)) # Adjusted figsize to account

# Line chart
plt.subplot(1,2,1) #charged to 1 row,3 columns,possible

plt.plot(months,
         sales,
         marker="O")

plt.title("sales Trend")

plt.xlabel("Months")

plt.ylabel("Sales")

# Bar chart
plt.subplot(1,3,2) # charged to 1 row,3 columns,possible

plt.bar(months,
        sales,
        color='orange')

plt.title("Sales comparison")

```

```python
import matplotlib.pyplot as plt

# Data
lables = ["A","B","C","D"]
sizes = [15,30,45,10]

# Create pie chart
plt.pie(sizes, labels=lables,
        autopct='%1.1f%%')

# Title
plt.title("Pie Chart Example")

# Show chart
plt.show()

```
