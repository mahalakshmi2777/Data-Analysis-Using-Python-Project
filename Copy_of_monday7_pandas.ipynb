```python
import pandas as pd

data = {
    'emp_nm' :['maha','Lakshmi'],
    'emp_sal':[300,500],
}
df = pd.DataFrame(data)
df.rename(columns={
    'emp_nm': 'Employee Name',
    'emp_sal':'Employee Salary'
    }, inplace = True)

print(df)


```

```python
import pandas as pd

data = {
    "Region":["South","North","East","West","South","North"],
    "Products":["Labtop","Mobile","Tablet","Laptop","Mobile","Tablet"],
     "Sales" :[6000,7000,8000,9000,10000,11000]

}
df = pd.DataFrame(data)
print(df)
```

```python
south_sales = df[df["Region"] =="South"]

print(south_sales)
```

```python
high_sal = df[df["sal"]>10000]

print(high_sal)
```

```python
sorted_df = df.sort_values("Sales")

print(sorted_df)
```
