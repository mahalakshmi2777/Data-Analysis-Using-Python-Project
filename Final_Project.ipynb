```python
import pandas as pd

# Employee dataset
data = {
    "Employee":["A","B","C","D"],
    "Department":["HR","Sales","None","IT"],
    "Salary": [3000,5000,7000,9000],
    "Attrition": ["No","Yes","No","Yes"]
}

df = pd.DataFrame(data)

#Total employees left
left_employees = df[df["Attrition"] == "Yes"]

print("Total employees left:")
print(left_employees)

# Attrition by Department
print("\nAttrition by Department:")
print(left_employees.groupby("Department").size())
```

```python
dept_attrition = df.groupby("Department")["Attrition"].value_counts()

print("\nAttrition by Department:")
print(dept_attrition)
```

```python
salary_attrition = df.groupby("Salary")["Attrition"].value_counts()

print("\nAttrition by Salary:")
print(salary_attrition)
```

```python
attrition_count = df["Attrition"].value_counts()

print("\nAttrition Count:")
print(attrition_count)
```

```python
import matplotlib.pyplot as plt

attrition_count = df["Attrition"].value_counts()
df["Attrition"].value_counts().plot(kind="bar")
plt.title("Employee Attrition")
plt.show()

plt.pie(
    attrition_count,
    labels=attrition_count.index,
    autopct='%1.1f%%'
)

plt.title("Employee Attrition")
plt.show()
```

```python
avg_salary_by_dept = df.groupby("Department")["Salary"].mean()

print("\nAverage Salary by Department:")
print(avg_salary_by_dept)
```

```python
summary = df.groupby("Department").agg({
    "Salary": "mean",
    "Attrition": "value_counts"
})

print("\nSummary:")
print(summary)
```
