```python
import pandas as pd

data = {
       "Name" : ["Maha","Lakshmi","Ravi"],
       "Sales":[500,1500,2500]
}
df = pd.DataFrame(data)
print(df.duplicated())

```

```python
import pandas as pd

data = {
    'emp_nm' :['maha','Lakshmi'],
    'emp_sal':[300,500],
}
df = pd.DataFrame(data)
df.rename(columns={
    'emp_nm': 'Employee Name',
    'emp_sal':'Employee Salary'
    }, inplace = True)

print(df)


```
